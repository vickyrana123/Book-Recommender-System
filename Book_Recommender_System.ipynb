```python
import numpy as np
import pandas as pd
```

```python
books=pd.read_csv('BX-Books.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
books.head()
```

```python
books=books[['ISBN','Book-Title', 'Book-Author', 'Year-Of-Publication', 'Publisher']]
books.head(5)
```

```python
books.rename(columns={'Book-Title':'title','Book-Author':'author','Year-Of-Publication':'Year','Publisher':'publisher'},inplace=True)
```

```python
books.rename(columns={'Year':'year'},inplace=True)
books.head(5)
```

```python
users=pd.read_csv('BX-Users.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
users.head(5)
```

```python
users.rename(columns={'User-ID':'user-id','Location':'location','Age':'age'},inplace=True)
users.head(5)
```

```python
ratings=pd.read_csv('BX-Book-Ratings.csv',sep=';',error_bad_lines=False,encoding='latin-1')
```

```python
ratings.head(5)
```

```python
ratings.rename(columns={'User-ID':'user-id','Book-Rating':'rating'},inplace=True)
ratings.head(5)
```

```python
books.shape
```

```python
users.shape
```

```python
ratings.shape
```

```python
x=ratings['user-id'].value_counts()>200
y=x[x].index
```

```python
y
```

```python
ratings=ratings[ratings['user-id'].isin(y)]
```

```python
ratings.shape
```

```python
rating_with_books=ratings.merge(books,on='ISBN')
rating_with_books.shape
```

```python
number_rating=rating_with_books.groupby('title')['rating'].count().reset_index()
```

```python
number_rating.rename(columns={'rating':'number of ratings'},inplace=True)
number_rating
```

```python
final_rating=rating_with_books.merge(number_rating,on='title')
final_rating.shape
```

```python
final_rating[final_rating['number of ratings']>=50]
```

```python
final_rating.drop_duplicates(['user-id','title'],inplace=True)
final_rating.shape
```

```python
book_pivot=final_rating.pivot_table(columns='user-id',index='title',values='rating')
```

```python
book_pivot.fillna(0,inplace=True)
```

```python
from scipy.sparse import csr_matrix
book_sparse=csr_matrix(book_pivot)
```

```python
type(book_sparse)
```

```python
from sklearn.neighbors import NearestNeighbors
model=NearestNeighbors(algorithm='brute')
```

```python
model.fit(book_sparse)
```

```python
distances, suggestions=model.kneighbors(book_pivot.iloc[237,:].values.reshape(1,-1),n_neighbors=6)
```

```python
suggestions
```

```python
for i in range(len(suggestions)):
  print(book_pivot.index[suggestions[i]])
```

```python
book_pivot.index[14]
```

```python
def recommend_book(book_name):
  book_id=np.where(book_pivot.index==book_name)[0][0]
  distances,suggestions=model.kneighbors(book_pivot.iloc[book_id,:].values.reshape(1,-1))

  for i in range(len(suggestions)):
    if i==0:
      print("The suggestions for",book_name,"are:")
    if not i:
      print(book_pivot.index[suggestions[i]])
```

```python
recommend_book('Animal Farm')
```
